Pressure	p	(and	density	$\rho$)	decreases exponentially with height: 
$$p = p_0 e^{\frac{-z}{H}}$$

```python
import matplotlib.pyplot as plt
import numpy as np

H = 8000
dz = 1000
p0 = 1000


z = np.arange(0,10000,dz)
def p(x):
    pres = p0*np.exp(-x/H)
    return pres

plt.plot(z,p(z), 'blue')
plt.title('Vertical Pressure')
plt.xlabel('Dz (m)')
plt.ylabel('P(z) ')

plt.grid(True)
plt.legend(['p*exp(-z/H)'], loc='upper right')

## Show the graph
plt.show()
```

```python
g = 9.8 #m/s^2 const
z_bkw = (-1/g*(p(z)-p(z-dz)))/dz      #backward
z_frw = (-1/g*(p(z+dz)-p(z)))/dz      #forward
z_ctw = (-1/g*(p(z+dz)-p(z-dz)))/(2*dz)    #central
z_anl = p0/(g*H)*np.exp(-z/H)        #analytical

#using latex in label
plt.rc('text', usetex= True)

#plot
plt.plot(z,z_bkw, color = 'blue', label = 'Backward')
plt.plot(z,z_frw, color = 'red', label = 'Forward')
plt.plot(z,z_ctw, color = 'green', label = 'Central')
plt.plot(z,z_anl, linestyle='dashed', marker='o', label = 'Analytical')

plt.xlabel('Dz(m)')
plt.ylabel('Density')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
