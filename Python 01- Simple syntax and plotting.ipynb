## Plot y = sin(x)

```python
import matplotlib.pyplot as plt
import numpy as np

nsample = 1000
f = 5
x = np.arange(nsample)
y = np.sin(2*np.pi*f*x/nsample)

plt.xlabel('x')
plt.ylabel('sinx')
plt.title('Diagram of SIN(X)')
plt.plot(x, y)

plt.show()
```

## Enter some numbers then calculate their mean, median, variance, std value

```python
import numpy as np

n = int(input("enter the number n: "))
a = np.zeros(n)
i = 0
while i<n:
    print("input number",i+1)
    a[i] = input()
    i=i+1
print (a)


print ("mean: ", np.mean(a))
print ("median: ", np.median(a))
print ("variance: ", np.var(a))
print ("standard variation: ", np.std(a))
```
